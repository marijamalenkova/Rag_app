```python
# generate_embeddings.py
# Run this ONCE locally: python generate_embeddings.py

import pickle
import numpy as np
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter


print("Chunking documents...")
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separators=["\n\n", "\n", ". ", " ", ""],
)
chunks = []
for doc in DOCUMENTS:
    chunks.extend(splitter.split_text(doc))

print(f"Created {len(chunks)} chunks. Embedding...")
model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectors = model.embed_documents(chunks)

print("Saving embeddings.pkl...")
with open("embeddings.pkl", "wb") as f:
    pickle.dump({
        "chunks": chunks,
        "vectors": np.array(vectors, dtype=np.float32)
    }, f)

print("Done! Commit embeddings.pkl to your repo.")
```
